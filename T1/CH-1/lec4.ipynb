```python
df1=pd.DataFrame({"sub":["FSD","python","DM"],
                  "fac":["nas","vha","mss"],
                 "id":[1,2,4]})
df2=pd.DataFrame({"id":[1,2,3],"mark":[9,10,8]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="id")
```

```python
pd.merge(df1,df2,on="id",how="inner")
```

```python
pd.merge(df1,df2,on="id",how="outer")
```

```python
df1.merge(df2,on="id")
```

```python
df1.merge(df2,on="id",how="left")
```

```python
df1.merge(df2,on="id",how="right")
```

```python
df1.merge(df2,how="cross")
```

```python
df1=pd.DataFrame({"sub":["FSD","python","DM"],
                  "fac":["nas","vha","mss"],
                 "id_x":[1,2,4]})
df2=pd.DataFrame({"id_y":[1,2,3],"mark":[9,10,8],"fac":["nas","vha","mss"]})
```

```python
df1.merge(df2,left_on="id_x",right_on="id_y")
```

```python
df1.merge(df2,left_index=True,right_index=True)
```

```python
df1.merge(df2,left_on="id_x",right_on="id_y",suffixes=("_df1","_df2"))
```

```python
# import like this
items_path = "items.csv"
q1_path = "quarter-1.csv"
q2_path = "quarter-2.csv"
q3_path = "quarter-3.csv"


q1= pd.read_csv(q1_path)
q2 = pd.read_csv(q2_path)
q3 = pd.read_csv(q3_path)

items = pd.read_csv(items_path)
```

```python
pd.concat((q1,q2,q3))
```

```python
df=pd.concat((q1,q2,q3),keys=["Q1","Q2","Q3"])
```

```python
df=df.reset_index()
```

```python
df
```

```python
items
```

```python
df=df.merge(items,on="item_id")
```

```python
df
```

```python
df["quantity"]
```

```python
df.groupby(["level_0","item_name"],as_index=False)["quantity"].sum()
```

```python
df.groupby(["level_0","item_name"],as_index=False)["quantity"].sum().sort_values("quantity",ascending=False).drop_duplicates(subset=["level_0"])
```

```python
df["price"]=df["item_price"].apply(lambda x: float(x[1:]))
```

```python
df.price
```

```python
df["totalprice"]=df["quantity"]*df["price"]
```

```python
df.totalprice
```

```python
df.groupby(["level_0","item_name"],as_index=False)["totalprice"].sum().sort_values("totalprice",ascending=False).drop_duplicates(subset=["level_0"])
```
