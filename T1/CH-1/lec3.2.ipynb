```python
def Grade(Total):
    if Total>300:
        return "A+"
    if Total>220:
        return "A"
    if Total>180:
        return "B"
    if Total>140:
        return "C"
    return "F"
```

```python
df["Grade"]=df["Total"].apply(Grade)
```
