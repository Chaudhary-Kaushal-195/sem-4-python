```python
Q1=df.horsepower.quantile(0.25)
```

```python
Q3=df.horsepower.quantile(0.75)
```

```python
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
h1=Q3+1.5*IQR
l1=Q1-1.5*IQR
h1,l1
```

```python
wo=df[(df["horsepower"]>l1)&(df["horsepower"]<h1)]
```

```python
sns.boxenplot(wo["horsepower"])
```
