```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default="notebook"
```

```python
px.histogram(data_frame=df,
             x="dep",
             y="salary",
             histfunc="avg",
             color="gender",
            )
```
