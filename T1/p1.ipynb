```python
#pb 139  The following dictionary shows how five people follow each other on Instagram:
#  instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
#  E.g., the list for person1 has the value on index 2 as 1 which means person1 followsperson3 and a directed edge should be 
# added from person1 to person3.
 
#  Using networkx library, create a directed graph.
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
instagram = {'person1': [0,1,1,0,1], 
             'person2': [0,0,1,0,1], 
             'person3': [1,1,0,1,1],
             'person4': [1,1,1,0,0], 
             'person5': [1,1,0,0,0]}

for i in instagram:
    for j in range(0,4):
        if instagram[i][j]==1:
            G.add_edge(i,j)
nx.draw(G,with_labels=True,node_size=3000,node_color="pink",width=1)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
p=list(instagram.keys())
for i,person in enumerate(p):
    for j,val in enumerate(instagram[person]):
        if val==1:
            G.add_edge(person,p[j])
nx.draw_networkx(G,node_size=2500,node_color="red",edge_color="blue",width=2)
plt.show()
```
