## PB139
#### The following dictionary shows how five people follow each other on Instagram:
#### instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4':[1,1,1,0,0],'person5'[1,1,0,0,0]}
#### E.g., the list for person1 has the value on index 2 as 1 which means person1 followsperson3 and a directed edge should be
#### added from person1 to person3

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
p=list(instagram.keys())
for i,person in enumerate(p):
    for j,val in enumerate(instagram[person]):
        if val==1:
            G.add_edge(person,p[j])
nx.draw_networkx(G,node_size=2500,node_color="red",edge_color="blue",width=2)
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({
 "StudyHours":[2,3,5,6,8],
 "Marks":[50,60,75,85,95],
 "Gender":["M","F","M","F","M"]
})
def get_grade(marks):
    if marks >= 85:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "D"
df["Grade"]=df["Marks"].apply(get_grade)
sns.scatterplot(data=df,x="StudyHours",y="Marks",style="Gender",hue="Grade")
```
